# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. 
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos 

You need to find following details: 

A)Rank

B) Name

C) Artist

D) Upload date

E) Views 

```python
driver = webdriver.Edge()

# URL 
url1 = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'

# use GET to open the page
response1 = requests.get(url1)

use soup to hold the html
soup1 = BeautifulSoup(response1.content, 'html.parser')

# create a variable to look for class and wikitable in html
table1 = soup1.find('table', {'class': 'wikitable'})

# create the variables to hold the lists
ranks = []
video_titles = []
channel_names = []
upload_dates = []
views = []

# use a for loop to assign data the lists
for row in table1.find_all('tr')[1:]:  
    cols = row.find_all('td') #look for td data in html
    if len(cols) > 1:  # compare the length of columns
        rank = cols[0].get_text(strip=True)  # add data to the columns using get_text to sort td html
        video_title = cols[1].get_text(strip=True)   
        channel_name = cols[2].get_text(strip=True)  
        upload_date = cols[4].get_text(strip=True)  
        view_count = cols[3].get_text(strip=True)  
        
        # Use append to add data to the lists
        ranks.append(rank)
        video_titles.append(video_title)
        channel_names.append(channel_name)
        upload_dates.append(upload_date)
        views.append(view_count)

# Create variable to hold list data in columns
data1 = {
    'Name': ranks,
    'Artist': video_titles,
    'Views': channel_names,
    'Upload Date': upload_dates,
    'Upload Dates': views
}

#create a variable to present the data1 in a DataFrame

df1 = pd.DataFrame(data1)

# Display the DataFrame
(df1)
```

# 4. Scrape the details of trending repositories on Github.com. 

Url = https://github.com/ 

You have to find the following details: 

A) Repository title 

B) Repository description 

C) Contributors count 

D) Language used 


```python
# setup driver for edge
driver = webdriver.Edge() 

# create the URL variable and get it opening full screen and wait to load
url4 = 'http://github.com/'
driver.get(url4)
driver.fullscreen_window()
time.sleep(15)  

try:
    # create the "Open Source" variable to wait for the button to be present
    open_source_svg = WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.XPATH, "//button[contains(@class, 'HeaderMenu-link') and contains(., 'Open Source')]"))
    )

    # use action chains to hover over variable button
    actions = ActionChains(driver)
    actions.move_to_element(open_source_svg).perform()
    # short rest to load the page
    time.sleep(5)  
    
    # identify the trending button to be present in the dropdown menu
    trending_button = WebDriverWait(driver, 30).until(
        EC.element_to_be_clickable((By.XPATH, "//a[contains(@href, '/trending')]"))
    )

    # Click the trending button
    trending_button.click()

    # short wait for the new page to load
    time.sleep(5)
    
    # create a variable to look through the html
    response4 = driver.page_source
    soup4 = BeautifulSoup(response4, 'html.parser')

    # create the repositories variable to hold the data and webscrape the html
    repositories = soup4.find_all('article', class_='Box-row')

    # for loop to contain all a class html, print the necessary data from repository, description, stars, forks, language and divide them
    for repo in repositories:
        title = repo.find('a', class_='Link').text.strip()
        description = repo.find('p', class_='col-9 color-fg-muted my-1 pr-4').text.strip()
        stars = repo.find('a', href=lambda href: href and "stargazers" in href).text.strip()
        forks = repo.find('a', href=lambda href: href and "forks" in href).text.strip()
        language = repo.find('span', itemprop='programmingLanguage').text.strip()

        print(f"Repository Title: {title}")
        print(f"Description: {description}")
        print(f"Stars: {stars}")
        print(f"Forks: {forks}")
        print(f"Language: {language}")
        print("-----")
    
except Exception as e:
    print(f"An error occurred: {e}")

driver.fullscreen_window()
driver.quit()
```

# 8. Details of Datasets from UCI machine learning repositories. 
Url = https://archive.ics.uci.edu/ 
You have to find the following details: 

A) Dataset name 

B) Data type 

C) Task 

D) Attribute type 

E) No of instances 

F) No of attribute G) Year

```python
# prepare the browser
driver = webdriver.Edge()  
# Open the URL and fullscreen
url8 = 'https://archive.ics.uci.edu/'
driver.get(url8)
driver.fullscreen_window()

# Wait for the "Datasets" link to be present, creating a variable to click the button
try:
    datasets_link = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.LINK_TEXT, "Datasets"))
    )
    datasets_link.click()
except Exception as e:
    print(f"An error occurred: {e}")
    

driver.fullscreen_window()

# create a block 
try:
    # create a variable for dataset names using the XPath
    dataset_elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.XPATH, '//h2/a'))
    )
    
    # Create variable to extract text from all found data elements
    dataset_names = [element.text for element in dataset_elements]


except Exception as e:
    print(f"An error occurred: {e}")

# try block that waits for the variable truncate to be visible and locates span.truncate by css
try:
    
    truncate_elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'span.truncate'))
    )
    
    # create a variable to store data extracted from span.truncate
    truncate_texts = [element.text for element in truncate_elements]


except Exception as e:
    print(f"An error occurred while fetching truncate elements: {e}")
# create the dataframe variable using found data lists, as most data is in span the for loop divides the elements
# in order.
data8 = []
for i in range(0, len(truncate_texts), 4):
    task = truncate_texts[i]
    data_type = truncate_texts[i + 1]
    instances = truncate_texts[i + 2]
    features = truncate_texts[i + 3]
     # create the variable to organise the data
    dataset_name = dataset_names[i // 4]  
    data8.append([dataset_name, task, data_type, instances, features])

# Create the DataFrame with the new column headers
df8 = pd.DataFrame(data8, columns=['Dataset', 'Task', 'Data Type', 'Instances', 'Features'])

# Display the DataFrame
print(df8)

driver.quit()
```
